```python
f = open('100+100_with_mask_image_train_val_FN_metric.txt','r')
train_dice = []
for line in f:
    train_dice.append(float(line[:-1]))
print(train_dice)
f.close()

print('ori FN: ',min(train_dice))
```

```python
f = open('100+100_with_mask_image_train_val_FP_metric.txt','r')
val_loss = []
for line in f:
    val_loss.append(float(line[:-1]))
print(val_loss)
f.close()

print('ori FP: ',min(val_loss))
```

```python
f = open('100_with_mask_image_train_loss.txt','r')
train_loss = []
for line in f:
    train_loss.append(float(line[:-1]))
print(train_loss)
f.close()

print('ori: ',max(train_loss))
```

# Plot dice and loss

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import seaborn as sns
sns.set_style("white")

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
#plt.plot(history.history['my_iou_metric'][1:]) # my_dice metric
plt.plot(train_dice)
#plt.plot(history.history['val_my_iou_metric'][1:]) # val_my_dice_metric
plt.plot(val_dice)
plt.ylabel('dice')
plt.xlabel('epoch')
plt.legend(['train','Validation'], loc='upper left')

plt.title('model Dice') #model Dice

plt.subplot(1,2,2)
#plt.plot(history.history['loss'][1:])
#plt.plot(history.history['val_loss'][1:])
plt.plot(train_loss)
plt.plot(val_loss)
plt.ylabel('val_loss')
plt.xlabel('epoch')
plt.legend(['train','Validation'], loc='upper left')
plt.title('model loss')
```
